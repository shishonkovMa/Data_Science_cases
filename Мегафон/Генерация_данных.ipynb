# Генерация данных
## Абоненты

Таблица "Абоненты/Пользователи" из задания на Stepik - наш шаблон

```python
import numpy as np
import pandas as pd
import random
```

Сделаем возможным ввод с клавиатуры количества пользователей и количества событий

```python
n = int(input('Количество пользователей: '))
nm = int(input('Количество событий: '))
```

Также определим названия тарифов в переменной "name_t" и названия возможных городов в переменной "cities".

```python
col_1 = ["id", "Текущий_баланс","Дата_добавления","Возраст",
         "Город_проживания","Временная_метка_последней_активности", "Активный_тариф"]

name_t = ['БезПереплат','Максимум','Минимум','Лучший','ДажеПриНоле','Активный','Безлимит','Про']
m = len(name_t)

df_1 = pd.DataFrame(columns = col_1)

df_1['id'] = [i for i in range(1, n+1)]
df_1["Текущий_баланс"] = [round(random.uniform(0, 5000),1) for i in range(n)]
df_1['Дата_добавления'] = pd.to_datetime(['{}-{}-{}'
                                          .format(random.randrange(1998, 2023), 
                                                  random.randrange(1, 13),
                                                  random.randrange(1, 29)) for i in range(n)])
df_1['Возраст'] = [round(random.randrange(14, 87)) for i in range(n)]
cities = ["Москва","Санкт-Петербург","Красноярск","Суздаль","Новороссийск","Кодинск", "Владивосток",
          "Мурманск","Нижний Новгород","Анапа","Сочи","Белгород","Казань","Новосибирск","Крымск","Новочеркасск"]
df_1['Город_проживания'] = [cities[random.randrange(0, len(cities))] for i in range(n)]
df_1['Временная_метка_последней_активности'] = pd.to_datetime(['2022-{}-{} {}:{}:{}'
                                          .format(random.randrange(1, 4),
                                                  random.randrange(1, 29),
                                                 random.randrange(0, 24),
                                                 random.randrange(0, 60),
                                                 random.randrange(0, 60)) for i in range(n)])
df_1['Активный_тариф'] = [random.randrange(1, m+1) for i in range(n)]
df_1
```

## Тарифы

Таблица "Тарифы" из задания на Stepik - наш шаблон

```python
col_2 = ["id", 'Название','Дата_начала_действия','Дата_конца_действия','Объем_минут','Объем_смс','Объем_трафика_мб']
df_2 = pd.DataFrame(columns = col_2)


df_2['id'] = [i for i in range(1, m+1)]

df_2['Название'] = [i for i in name_t]
df_2['Дата_начала_действия'] = pd.to_datetime(['{}-{}-{}'
                                          .format(random.randrange(2018, 2023), 
                                                  random.randrange(1, 13),
                                                  random.randrange(1, 29)) for i in range(m)])
df_2['Дата_конца_действия'] = pd.to_datetime(['{}-{}-{}'
                                          .format(random.randrange(2023, 2029), 
                                                  random.randrange(1, 13),
                                                  random.randrange(1, 29)) for i in range(m)])
df_2['Объем_минут'] = [random.randrange(100, 801)//100 * 100 for i in range(1, m+1)]
df_2['Объем_смс'] = [random.randrange(50, 201)//10 * 10 for i in range(1, m+1)]
df_2['Объем_трафика_мб'] = [random.randrange(1024, 16385)//10 * 10 for i in range(1, m+1)]
df_2
```

## События

Таблица "События" из задания на Stepik - наш шаблон

```python
col_3 = ['id','Метка_времени','id_абонента','Тип_услуги_звонок_смс_трафик','Объем_затраченных_единиц']
df_3 = pd.DataFrame(columns = col_3)

df_3['id'] = [i for i in range(1, nm+1)]
df_3['Метка_времени'] = pd.to_datetime(['2022-03-{} {}:{}:{}'
                                          .format(random.randrange(1, 29),
                                                 random.randrange(0, 24),
                                                 random.randrange(0, 60),
                                                 random.randrange(0, 60)) for i in range(nm)])
df_3['id_абонента'] = [random.randrange(1, n+1) for i in range(1, nm+1)]
types = ['Звонок','Смс','Интернет']
df_3['Тип_услуги_звонок_смс_трафик'] = [types[random.randrange(0, len(types))] for i in range(nm)]
for i in range(nm):
    if df_3['Тип_услуги_звонок_смс_трафик'][i] == 'Звонок':
        df_3['Объем_затраченных_единиц'][i] = random.randrange(45, 801)
    elif df_3['Тип_услуги_звонок_смс_трафик'][i] == 'Смс':
        df_3['Объем_затраченных_единиц'][i] = random.randrange(0, 201)
    elif df_3['Тип_услуги_звонок_смс_трафик'][i] == 'Интернет':
        df_3['Объем_затраченных_единиц'][i] = random.randrange(16385)
df_3
```

Сохраним получившиеся датасеты в файлах.

```python
df_1.to_csv('abonent.csv')
df_2.to_csv('tariff.csv')
df_3.to_csv('events.csv')
```
